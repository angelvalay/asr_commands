```python
import numpy as np
import librosa.display
import librosa.output
import matplotlib.pyplot as plt
data, sr = librosa.load('data/data_with_roise/Command_Avanza_30.wav',sr=44100)
```

```python
plt.figure(figsize=(14,5))
librosa.display.waveplot(data, sr=sr)
```

```python
_noise = np.array([])
_data = np.array([])
_flag_start = False

for d in data:
    if _flag_start:
        _data = np.append(_data, [d])
    else:
        _noise = np.append(_data, [d])
    if d > 0.08 and not _flag_start:
        _flag_start = True
```

```python
plt.figure(figsize=(14,5))
librosa.display.waveplot(_data, sr=sr)
```

```python
plt.figure(figsize=(14,5))
librosa.display.waveplot(_data, sr=sr)
librosa.output.write_wav('test_avanza_30.wav', _data, sr)
```

```python
_n_data = np.array([])
_flag_start = False
for item in reversed(_data):
    if _flag_start:
        _n_data = np.append(_n_data, [item])
    if item > 0.08 and not _flag_start:
        _flag_start = True
_n_data = _n_data[::-1]
```

```python
plt.figure(figsize=(14,5))
librosa.display.waveplot(np.asfortranarray(_n_data), sr=sr)
librosa.output.write_wav('test_avanza_30.wav', np.asfortranarray(_n_data), sr)


```
