```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#load data
data_features = pd.read_csv('../data/features_voice.csv', thousands=' ')
# data_features.head()
```

```python
data = data_features.groupby(by='label')
data
```

```python
data_features.plot(kind='scatter', x='spectral_centroid', y='rolloff')
#show data spectral_centroid-rolloff
# data_features.plot(kind='scatter', x='spectral_centroid', y='rolloff')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
for key, grp in data_features.groupby(['label']):
    ax.plot(grp['spectral_centroid'], grp['chroma_stft'], label=key)

ax.legend()
plt.show()
```

```python
#show data spectral_centroid-chroma_sftf
data_features.plot(kind='scatter', x='spectral_centroid', y='zero_crossing_rate')
plt.show()
```

```python
corr_matrix=data_features.corr()
```

```python
corr_matrix['chroma_stft'].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix
attributes=['chroma_stft','spectral_bandwidth', 'rolloff','zero_crossing_rate','spectral_centroid']
scatter_matrix(corr_matrix[attributes], figsize=(12,8))
# ax =sns
plt.show()
```
